```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

def cluster_based_undersampling(data, n_clusters=10, target_proportion=0.5, random_state=42):
    # Separate the majority and minority class data
    majority_data = data[data['employment_status'] == 0].copy()
    minority_data = data[data['employment_status'] == 1].copy()

    # Prepare data for clustering
    scaler = StandardScaler()
    majority_features = scaler.fit_transform(majority_data.drop('employment_status', axis=1))

    # Clustering the majority data
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
    majority_data['cluster'] = kmeans.fit_predict(majority_features)

    # Determine number of samples to take from each cluster
    total_majority_samples = int(len(minority_data) * target_proportion)
    samples_per_cluster = total_majority_samples // n_clusters

    # Sampling representatives from each cluster
    sampled_majority_data = majority_data.groupby('cluster').apply(
        lambda x: x.sample(n=min(samples_per_cluster, len(x)), random_state=random_state)
    ).reset_index(drop=True)

    # Combine sampled majority data with minority data
    undersampled_data = pd.concat([sampled_majority_data, minority_data], axis=0)
    undersampled_data = undersampled_data.sample(frac=1, random_state=random_state).reset_index(drop=True)

    return undersampled_data



# Apply the modified undersampling function
undersampled_data = cluster_based_undersampling(df1, n_clusters=10, target_proportion=1.0)  # 1:1 ratio

# Check new balance
print(undersampled_data['employment_status'].value_counts())
```

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

def plot_learning_curve(estimator, X, y, title="Learning Curve", cv=None, n_jobs=None, train_sizes=np.linspace(0.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    plt.xlabel("Training examples")
    plt.ylabel("Score")

    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes, scoring='accuracy')

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt

# Use logistic regression from sklearn for the learning curve
logistic_regression_model = LogisticRegression(max_iter=10000)

# Plot learning curve
plot_learning_curve(logistic_regression_model, X_train, y_train, cv=5, n_jobs=-1)
plt.show()
```
